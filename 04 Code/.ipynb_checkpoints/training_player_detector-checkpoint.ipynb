# Retreiving Data

```python
!pip install roboflow
from roboflow import Roboflow

os.chdir("03 Data")
DATA_dir = os.getcwd()
print(DATA_dir)

rf = Roboflow(api_key="INSERT_API_KEY_HERE")
project = rf.workspace("roboflow-jvuqo").project("football-players-detection-3zvbc")
version = project.version(20)
dataset = version.download("yolov8")
```

```python
# Update data.yaml so YOLO uses the correct relative paths for train and val images
!sed -i '' 's|\(train: \).*|\1../train/images|' "{dataset.location}/data.yaml"
!sed -i '' 's|\(val: \).*|\1../valid/images|' "{dataset.location}/data.yaml"
# sometime i get issues with this chunk. if you do ensure that the numpy version < 2.0.0  and rerun
```

# Training

```python
yaml_path = os.path.join(dataset.location, "data.yaml")
print("YAML file exists:", os.path.exists(yaml_path))
```

```python
with open(yaml_path, "r") as f:
    print(f.read())
```

```python
train_dir = os.path.join(dataset.location, "../football-players-detection-20/train/images")
val_dir = os.path.join(dataset.location, "../football-players-detection-20/valid/images")

print("Train images:", len(os.listdir(train_dir)))
print("Validation images:", len(os.listdir(val_dir)))
```

```python
!yolo task=detect mode=train model=yolov8s.pt data="{dataset.location}/data.yaml" batch=6 epochs=25 imgsz=1280 plots=True
# by defalut this model trains with 640x640 resolution. This default setting reduces our pixels and makes ball detection difficult because the ball already takes a few pixels.
# Therefore, I increased the input resolution from to 1280x1280 which increased accuracy for ball detection but made training slower.

# I set the batch size to 6 because i already slowed down the model by increasing the input resolution so i want to pick a smaller batch to speed up training

# originally I wanted to use model=yolov8x.pt but it took way too long. I also decreased the number of epochs to 25 to save time and make it less computationally expensive
```

```python
!ls "{DATA_dir}/runs/detect/train/"
```

```python
from IPython.display import Image
%cd {DATA_dir}
Image(filename=f'{DATA_dir}/runs/detect/train/confusion_matrix.png', width=600)
```

- For the most part the confusion matrix looks decent.
- It looks like the model is struggling with detecting the ball and half the time confuses ball with the background
- there are issues with goal keepers. the model also mis-identifies the background from time to time as goalkeepers
- the model does a good job with detecting players and referees
- sometimes other classes get falsely detecting instead of the model detecting the pixels as background

```python
%cd  {DATA_dir}
Image(filename=f'{DATA_dir}/runs/detect/train/results.png', width=600)
```

- train/box_loss: The model is effectively learning to localize objects, as indicated by the decreasing box loss.
- train/cls_loss: The model is improving its ability to correctly classify objects into the right categories.
- train/dfl_loss: The model is refining its bounding box edge predictions, showing better precision in object boundaries.
- metrics (precision and recall): Both precision and recall are increasing over epochs, meaning the model is becoming better at detecting true positives while reducing false positives and false negatives.
- Validation losses (val/box_loss, val/cls_loss, val/dfl_loss): These losses are also decreasing, suggesting that the model is generalizing well to unseen data. When compared to the training losses, there is no clear indication of overfitting.
- metrics/mAP50 and metrics/mAP50-95: Both metrics are increasing steadily, which is a positive sign that the modelâ€™s overall detection accuracy is improving across different IoU thresholds.
